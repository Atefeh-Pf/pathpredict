**PathPredict: Educational Success Forecasting**

## Goal
Since SOEP Practice Dataset doesn't contain geographic identifiers, we'll create **synthetic regional clusters** based on socioeconomic characteristics.

## Methodology
1. Select features that correlate with regional characteristics
2. Apply K-Means clustering to group similar individuals
3. Profile each cluster
4. Map clusters to real regional average data

## Why This Works
- People with similar socioeconomic profiles tend to live in similar regions
- This is the basis of **geodemographic segmentation** (widely used in marketing & social research)
- Our clusters serve as proxies for regional contexts

## 3. Determine Optimal Number of Clusters

We'll use the **Elbow Method** to find the optimal k.

We're targeting **16 clusters** (similar to Germany's 16 Bundesländer), but let's validate this choice.

### Ex-Course: Elbow Method

**Wie würden Cluster-Zentren berechnet:**

Beispiel mit 6 Personen, k=2 Gruppen:

**Personen:**
1. Person A: Einkommen=20k, Bildung=10
2. Person B: Einkommen=25k, Bildung=11
3. Person C: Einkommen=22k, Bildung=10
4. Person D: Einkommen=50k, Bildung=16
5. Person E: Einkommen=55k, Bildung=17
6. Person F: Einkommen=48k, Bildung=15

**Schritt 1: Zufällige Start-Zentren wählen**
- Zentrum 1: Einkommen=25k, Bildung=11
- Zentrum 2: Einkommen=50k, Bildung=16

**Schritt 2: Jede Person zum nächsten Zentrum zuordnen**

Abstand berechnen: 

√((Einkommen_Person - Einkommen_Zentrum)² + (Bildung_Person - Bildung_Zentrum)²)

- Person A zu Zentrum 1: √((20-25)² + (10-11)²) = √26 = 5.1
- Person A zu Zentrum 2: √((20-50)² + (10-16)²) = √936 = 30.6
- Person A gehört zu Zentrum 1 (näher)

Ergebnis:
- Cluster 1: A, B, C
- Cluster 2: D, E, F

**Schritt 3: Neue Zentren berechnen (Durchschnitt)**

Neues Zentrum 1:
- Einkommen = (20+25+22)/3 = 22.3k
- Bildung = (10+11+10)/3 = 10.3

Neues Zentrum 2:
- Einkommen = (50+55+48)/3 = 51k
- Bildung = (16+17+15)/3 = 16

**Schritt 4: Wiederhole Schritt 2-3**

Bis Zentren sich nicht mehr aendern!

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Konvertiere kategorische Spalten zu numerisch
df['gesund_org'] = pd.to_numeric(df['gesund_org'].astype(str).str.extract('(\d+)')[0], errors='coerce')
df['lebensz_org'] = pd.to_numeric(df['lebensz_org'].astype(str).str.extract('(\d+)')[0], errors='coerce')

# Elbow Code
clustering_features = ['einkommenj1', 'bildung', 'anz_kind', 'anz_pers', 'gesund_org', 'lebensz_org']
X = df[clustering_features].fillna(df[clustering_features].median())
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertias = []
K_range = range(2, 25)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, 'bo-', linewidth=2)
plt.axvline(x=16, color='red', linestyle='--', linewidth=2, label='k=16')
plt.xlabel('Anzahl Cluster (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Balkendiagramm statt Linie
plt.figure(figsize=(12, 6))
plt.bar(K_range, inertias, color='steelblue', edgecolor='black')
plt.axvline(x=16, color='red', linestyle='--', linewidth=3, label='Elbow bei k=16')
plt.xlabel('Anzahl Cluster (k)')
plt.ylabel('Inertia (Fehler)')
plt.title('Elbow Method - Balkendiagramm')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Zeige Zahlen
print("k=2:  Inertia =", int(inertias[0]))
print("k=10: Inertia =", int(inertias[8]))
print("k=16: Inertia =", int(inertias[14]), "← Elbow")
print("k=24: Inertia =", int(inertias[22]))
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Prepare data
X = df[available_features].copy()
X = X.fillna(X.median())  # Fill missing values with median

# Standardize features (important for K-Means!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f'✓ Prepared {X_scaled.shape[0]} observations with {X_scaled.shape[1]} features')
```

```python
# Elbow method: Test different numbers of clusters
inertias = []
K_range = range(2, 25)

print('Testing different cluster numbers (this may take 1-2 minutes)...')
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    if k % 5 == 0:
        print(f'  Tested k={k}')

print('✓ Elbow analysis complete')
```

```python
# Plot elbow curve
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
ax.axvline(x=16, color='red', linestyle='--', linewidth=2, label='Target: 16 clusters')
ax.set_xlabel('Number of Clusters (k)', fontsize=12)
ax.set_ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
ax.set_title('Elbow Method: Determining Optimal Number of Clusters', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

plt.tight_layout()
plt.show()

print('\n📊 Interpretation:')
print('The "elbow" is where the curve starts to flatten.')
print('k=16 is reasonable for our purpose (representing 16 German states).')
```

## 4. Create 16 Regional Clusters

Now we'll create our synthetic regions using K-Means with k=16.

```python
# Initialize cluster mapper
mapper = RegionalClusterMapper(n_clusters=16, random_state=42)

# Create clusters
df = mapper.create_clusters(df, available_features)
```

```python
# Check cluster distribution
print('Cluster sizes:')
cluster_counts = df['region_cluster'].value_counts().sort_index()
print(cluster_counts)

# Visualize distribution
fig, ax = plt.subplots(figsize=(12, 6))
cluster_counts.plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
ax.set_xlabel('Cluster ID', fontsize=12)
ax.set_ylabel('Number of Observations', fontsize=12)
ax.set_title('Distribution of Observations Across Clusters', fontsize=14, fontweight='bold')
ax.axhline(y=df.shape[0]/16, color='red', linestyle='--', label=f'Average ({df.shape[0]//16})')
ax.legend()
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 5. Profile Each Cluster

Let's understand the characteristics of each cluster.

```python
# Create cluster profiles
profile_vars = available_features + ['high_education'] if 'high_education' in df.columns else available_features

cluster_profiles = mapper.profile_clusters(df, profile_vars)

# Display profiles
print('\n📊 Cluster Profiles (Mean Values):')
print('='*80)
cluster_profiles.round(2)
```

## 7. Map Clusters to Real Regional Data

Now the crucial step: We map our synthetic clusters to **real average data** from German regions.

This data comes from:
- **Destatis** (Federal Statistical Office)
- **INKAR** (Regional Indicators)
- **KMK** (Education reports)

```python
# Load real regional data
regional_data = create_regional_data_mapping()

print('✓ Loaded regional data mapping')
print(f'\nRegional indicators available:')
print(regional_data.columns.tolist())

# Display sample
print('\n📊 Sample of Regional Data:')
regional_data.head(8)
```

```python
# Merge regional data with our dataset
print(f'Before merge: {df.shape}')

df_with_regions = df.merge(regional_data, on='region_cluster', how='left')

print(f'After merge: {df_with_regions.shape}')
print(f'✓ Successfully added {len(regional_data.columns)-1} regional indicators')

# Check for missing values
missing_regional = df_with_regions[regional_data.columns].isnull().sum()
if missing_regional.sum() == 0:
    print('✓ No missing values in regional data')
else:
    print('⚠️ Some missing regional values detected')
```

```python
# Display sample with regional data
sample_cols = ['id', 'bildung', 'einkommenj1', 'region_cluster', 
               'unemployment_rate', 'edu_spending_per_student', 'region_type']
available_sample_cols = [c for c in sample_cols if c in df_with_regions.columns]

print('Sample data with regional indicators:')
df_with_regions[available_sample_cols].head(10)
```

## 8. Analyze Regional Patterns

Let's explore how regional factors relate to educational outcomes.

```python
# Educational attainment by region type
if 'high_education' in df_with_regions.columns and 'region_type' in df_with_regions.columns:
    
    edu_by_region = df_with_regions.groupby('region_type')['high_education'].agg([
        ('Count', 'count'),
        ('Abitur_Rate', 'mean'),
        ('Avg_Education_Years', lambda x: df_with_regions.loc[x.index, 'bildung'].mean())
    ]).round(3)
    
    print('\n📊 Educational Outcomes by Region Type:')
    print('='*60)
    print(edu_by_region)
    
    # Visualize
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Abitur rate by region type
    edu_by_region['Abitur_Rate'].plot(kind='bar', ax=axes[0], color='coral', edgecolor='black')
    axes[0].set_ylabel('Proportion with Abitur+', fontsize=11)
    axes[0].set_xlabel('Region Type', fontsize=11)
    axes[0].set_title('Abitur Rate by Region Type', fontsize=12, fontweight='bold')
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')
    axes[0].set_ylim(0, 1)
    
    # Average education years
    edu_by_region['Avg_Education_Years'].plot(kind='bar', ax=axes[1], color='skyblue', edgecolor='black')
    axes[1].set_ylabel('Average Education Years', fontsize=11)
    axes[1].set_xlabel('Region Type', fontsize=11)
    axes[1].set_title('Average Education Years by Region Type', fontsize=12, fontweight='bold')
    axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')
    
    plt.tight_layout()
    plt.show()
```

```python
# Correlation: Regional factors vs Education
regional_indicators = ['unemployment_rate', 'edu_spending_per_student', 
                      'student_teacher_ratio', 'child_poverty_rate']

available_indicators = [i for i in regional_indicators if i in df_with_regions.columns]

if len(available_indicators) > 0 and 'bildung' in df_with_regions.columns:
    
    # Calculate correlations
    correlations = df_with_regions[available_indicators + ['bildung']].corr()['bildung'].drop('bildung')
    
    print('\n📊 Correlation: Regional Factors vs Education Years:')
    print('='*60)
    print(correlations.sort_values(ascending=False))
    
    # Visualize
    fig, ax = plt.subplots(figsize=(10, 6))
    correlations.sort_values().plot(kind='barh', ax=ax, color='teal', edgecolor='black')
    ax.set_xlabel('Correlation with Education Years', fontsize=12)
    ax.set_title('Regional Factors vs Educational Attainment', fontsize=14, fontweight='bold')
    ax.axvline(x=0, color='black', linewidth=0.8)
    plt.tight_layout()
    plt.show()
    
    print('\n💡 Interpretation:')
    print('- Positive correlation: Higher values associated with more education')
    print('- Negative correlation: Higher values associated with less education')
```

## 9. Create Summary Statistics by Cluster

```python
# Comprehensive cluster summary
summary_vars = ['bildung', 'einkommenj1', 'unemployment_rate', 
                'edu_spending_per_student', 'abitur_rate']
available_summary = [v for v in summary_vars if v in df_with_regions.columns]

cluster_summary = df_with_regions.groupby('region_cluster')[available_summary].agg(['mean', 'std']).round(2)

print('\n📊 Comprehensive Cluster Summary:')
print('='*80)
cluster_summary.head(16)
```

```python
# Save dataset with regional clusters and indicators
output_path = '../data/processed/soep_with_regions.csv'
df_with_regions.to_csv(output_path, index=False)

print(f'✓ Saved enhanced dataset to: {output_path}')
print(f'  - Total observations: {len(df_with_regions)}')
print(f'  - Total features: {len(df_with_regions.columns)}')
print(f'  - Regional indicators added: {len(regional_data.columns)-1}')

# Save cluster profiles
cluster_profiles.to_csv('../data/processed/cluster_profiles.csv')
print('\n✓ Saved cluster profiles to: data/processed/cluster_profiles.csv')
```

## 11. Validation: Does This Make Sense?

```python
# Sanity checks
print('🔍 VALIDATION CHECKS:\n')

# 1. Check if clusters are balanced
cluster_sizes = df_with_regions['region_cluster'].value_counts()
print(f'1. Cluster Balance:')
print(f'   Smallest cluster: {cluster_sizes.min()} observations')
print(f'   Largest cluster: {cluster_sizes.max()} observations')
print(f'   Ratio: {cluster_sizes.max()/cluster_sizes.min():.2f}x')
if cluster_sizes.max()/cluster_sizes.min() < 3:
    print('   ✓ Clusters are reasonably balanced')
else:
    print('   ⚠️ Some imbalance detected (acceptable for real-world data)')

# 2. Check if regional data was properly merged
print(f'\n2. Data Merge Quality:')
regional_cols = [c for c in regional_data.columns if c != 'region_cluster']
missing_in_merge = df_with_regions[regional_cols].isnull().sum().sum()
print(f'   Missing values in regional data: {missing_in_merge}')
if missing_in_merge == 0:
    print('   ✓ All regional data successfully merged')

# 3. Check correlation patterns
if 'bildung' in df_with_regions.columns and 'edu_spending_per_student' in df_with_regions.columns:
    corr_edu_spending = df_with_regions['bildung'].corr(df_with_regions['edu_spending_per_student'])
    print(f'\n3. Expected Correlation Pattern:')
    print(f'   Education vs Education Spending: {corr_edu_spending:.3f}')
    if corr_edu_spending > 0:
        print('   ✓ Positive correlation as expected')

print('\n' + '='*60)
print('✅ Synthetic regional clustering complete!')
print('='*60)
```
